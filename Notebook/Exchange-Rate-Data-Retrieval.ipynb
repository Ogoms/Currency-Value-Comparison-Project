```python
# !pip install requests
import requests
import pandas as pd
from pandas import json_normalize

api_url ='https://api.frankfurter.app/2013-01-01..2023-08-31'

response = requests.get(api_url)
# Preview the response
response.text

data = response.json()
# Convert the text response to a dictionary [also known as json]

```

```python
df = pd.DataFrame(data)

#Normalise the rates column
df_rates = json_normalize(df['rates'])

# df_rates
# Reset the index
df = df.reset_index()

# drop rates column
df = df.drop('rates', axis=1)


df_conc = pd.concat([df,df_rates], axis = 1)
df_conc 
df
```

```python
# Simple summary statistic and present any interesting conclusion that can be drawn
(
  df_conc
    # Please print out the descriptive statistics
    .describe().T
    
)
```

Here are some possible conclusions:

Variability of Exchange Rates: We can observe the range (min and max) and standard deviation of exchange rates for each currency pair. Higher standard deviation indicates higher variability in exchange rates, which could imply higher risk or volatility for traders or investors dealing with those currencies.

Central Tendency: The mean and median values provide insights into the central tendency of exchange rates. Comparing the mean and median can indicate if the data is skewed (if mean significantly differs from median). A significant difference could suggest outliers or asymmetry in the distribution of exchange rates.

Spread of Data: The interquartile range (IQR) provides information about the spread of exchange rates around the median. A larger IQR indicates a wider spread of data points, while a smaller IQR suggests a narrower spread.

Outliers: Examining the maximum and minimum values can help identify potential outliers in the data. Outliers could represent significant fluctuations or anomalies in exchange rates that may be of interest for further investigation.

Comparative Analysis: Comparing the descriptive statistics of different currency pairs can reveal interesting insights about the relative stability, volatility, or liquidity of various currencies in the dataset.

```python

import matplotlib.pyplot as plt


# Question 1: How has the exchange rate of the Euro (EUR) against the US Dollar (USD) changed over time?


# Check the structure of the response
print(data.keys())

# Extract dates and exchange rates
dates = list(data['rates'].keys())
eur_to_usd_rates = [data['rates'][date]['USD'] for date in dates]

# Create a DataFrame for the date
df_date = pd.DataFrame({'Date': dates, 'EUR_to_USD': eur_to_usd_rates})

# Convert Date column to datetime
df_date['Date'] = pd.to_datetime(df_date['Date'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_date['Date'], df_date['EUR_to_USD'], marker='o', linestyle='-')

# Adding labels and title
plt.title('EUR to USD Exchange Rate')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.grid(False)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Question 2: What is the distribution of exchange rates for the Chinese Yuan (CNY) against the Euro (EUR)
cny_to_eur_exchange_rate = df_conc['amount'] / df_conc['CNY']

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(cny_to_eur_exchange_rate, bins=20, edgecolor='black')
plt.title('Distribution of CNY to EUR Exchange Rate')
plt.xlabel('Exchange Rate')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# Question 3: Between the Great British pound (GBP) and the Japanese Yen (JPY), determine which country's currency is more volatile

# Calculate daily percentage change for GBP and JPY
df_conc['GBP_pct_change'] = df_conc['GBP'].pct_change()
df_conc['JPY_pct_change'] = df_conc['JPY'].pct_change()

# Compute standard deviation
gbp_std_dev = df_conc['GBP_pct_change'].std()
jpy_std_dev = df_conc['JPY_pct_change'].std()

# Compare standard deviations
if gbp_std_dev > jpy_std_dev:
    print("GBP is more volatile than JPY.")
elif gbp_std_dev < jpy_std_dev:
    print("JPY is more volatile than GBP.")
else:
    print("GBP and JPY have the same level of volatility.")

# Currency labels
currencies = ['GBP', 'JPY']

# Standard deviations
std_devs = [gbp_std_dev, jpy_std_dev]

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(currencies, std_devs, color=['blue', 'red'])
plt.title('Volatility Comparison: GBP vs JPY')
plt.xlabel('Currency')
plt.ylabel('Standard Deviation of Daily % Change')
plt.grid(axis='y')
plt.show()
```

```python
#Question 4: Is there a correlation between the exchange rates of the Canadian Dollar (CAD) and the American Dollar (USD)?

# Calculate correlation coefficient
correlation = df_conc['CAD'].corr(df_conc['USD'])

print(f'Correlation coefficient between CAD and USD exchange rates:, {correlation}')

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_conc['CAD'], df_conc['USD'], color='black', alpha=0.5)
plt.title('Scatter Plot: CAD to USD Exchange Rates')
plt.xlabel('CAD Exchange Rate')
plt.ylabel('USD Exchange Rate')
plt.grid(False)
plt.show()
```
